## Headline Season Numbers

Apexline accepted **79.1%** of all 2025 race laps as geometry-usable evidence against the oracle circuit line. The remaining bad laps are mostly telemetry-quality or pit-lap issues, with shape mismatches concentrated in a smaller set of circuits.

## Best Lap vs Average Of Five

| Averaging helps | Best RMSE | Avg-5 RMSE | Delta | Averaging hurts | Best RMSE | Avg-5 RMSE | Delta |
|---|---|---|---|---|---|---|---|
| Singapore Grand Prix | 13.5 m | 10.6 m | -2.9 m | Canadian Grand Prix | 5.0 m | 12.4 m | +7.4 m |
| Qatar Grand Prix | 4.9 m | 3.9 m | -1.0 m | Las Vegas Grand Prix | 7.1 m | 10.9 m | +3.9 m |
| Dutch Grand Prix | 8.2 m | 7.5 m | -0.7 m | Bahrain Grand Prix | 7.2 m | 9.9 m | +2.7 m |
| Belgian Grand Prix | 7.2 m | 6.5 m | -0.7 m | Saudi Arabian Grand Prix | 5.4 m | 7.5 m | +2.1 m |
| Spanish Grand Prix | 4.7 m | 4.2 m | -0.5 m | Abu Dhabi Grand Prix | 6.2 m | 7.5 m | +1.3 m |

Averaging is not automatically better. Canada is the sharpest counterexample: the best single validated lap fits at **5.0 m** RMSE, while the average of the five best fitted laps drifts out to **12.4 m** because racing-line variation pulls the mean away from the oracle line.
