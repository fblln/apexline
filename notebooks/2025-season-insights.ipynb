# Apexline 2025 Season Insights

This notebook is the deeper companion to the repo README. It focuses on what the 2025 full-season run teaches about lap suitability, shape validation, and compact oracle polylines.

Apexline accepted **86.3%** of all 2025 race laps as geometry-usable evidence against the oracle circuit line. Every fitted lap passed the visually calibrated shape limits; remaining rejects are telemetry-quality, pit-lap, path-length, or sample-count exclusions.

Evidence yield and shape agreement are different signals. Britain and Australia lose many laps to FastF1 accuracy metadata without showing a circuit-shape problem. The geometry fit does not invent a failure where the rendered traces match.


## Shape-Validation Result

All **23,042** laps that reached shape fitting passed. This is a measured result rather than a disabled check: each lap still receives RMSE and p95 metrics, but the 1.6% / 2.5% circuit-length limits match the gray-vs-color SVG audit.

The 3,647 rejected laps fail independent pre-fit evidence checks, so downstream consumers can distinguish telemetry suitability from geometric agreement.

Averaging is not automatically better. Canada is the sharpest counterexample: the best single validated lap fits at **5.0 m** RMSE, while the average of the five best fitted laps drifts out to **12.4 m** because racing-line variation pulls the mean away from the oracle line.
